```python
import numpy as np
import matplotlib.pyplot as plt 

#зафиксируем N=100000
P = np.array([1, 2, 4, 8, 16])

T1 = np.array([1.389960,  1.349377, 1.314956, 1.322059, 1.354956, 1.311884, 1.311301, 1.324456, 1.336677, 1.330311])
T1_mean = np.mean(T1)

T2 = np.array([1.341728, 1.362175, 1.341081, 1.349307, 1.320524, 1.299015, 1.318125, 1.304329, 1.309794, 1.307560])
T2_mean = np.mean(T2)

T4 = np.array([1.309199, 1.330780, 1.323265, 1.330528, 1.296882, 1.312641, 1.330167, 1.317230, 1.301944, 1.308928])
T4_mean = np.mean(T4)

T8 = np.array([1.314203, 1.303581, 1.319223, 1.317459, 1.301067, 1.312045, 1.291813, 1.321671, 1.325097, 1.318885])
T8_mean = np.mean(T8)

T16 = np.array([1.324920, 1.324596, 1.308695, 1.304184, 1.323065, 1.296354, 1.310657, 1.288465, 1.313674, 1.316860, 1.329041])
T16_mean = np.mean(T16)


T = np.array([T1_mean, T2_mean, T4_mean, T8_mean, T16_mean])

S = T[0] / T
E = S / P

plt.figure(figsize=(10, 6))
plt.plot(P, T, color='green')
plt.title("$T(P)$", size=14)
plt.xlabel("P - number of threads")
plt.ylabel("T - seconds")
plt.show()


plt.figure(figsize=(10, 6))
plt.plot(P, S, color='green')
plt.title("S(P)", fontsize=14)
plt.xlabel("P - number of threads")
plt.ylabel("S - acceleration")
plt.show()


plt.figure(figsize=(10, 6))
plt.plot(P, E, color='green')
plt.title("E(P)", fontsize=14)
plt.xlabel("P - number of threads")
plt.ylabel("E - efficiency")
plt.show()
```
